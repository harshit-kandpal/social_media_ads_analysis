# Feature Extraction

```python
clicked_data.columns
```

```python
feature =clicked_data[ ['Daily Time Spent on Site','Age', 'Area Income','Daily Internet Usage','Male']]

x = np.asarray(feature)
y = np.asarray(clicked_data['Clicked on Ad'])
```

```python
x[0:5]
```

Creating testing and training dataset

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=5)
```

# Machine learning model 

```python
from sklearn import svm

classifier1 = svm.SVC(kernel='linear',gamma='auto',C=2)
```

Training model

```python
classifier1.fit(x_train,y_train)
```

Testing model

'Daily Time Spent on Site','Age', 'Area Income','Daily Internet Usage','Male'

```python
classifier1.predict(np.asarray([[70 ,42,100000,500,1]]))
```

```python
classifier1.predict(np.asarray([[70 ,20,30000,200,0]]))
```

```python
y_predict1 = classifier1.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict1))
```
